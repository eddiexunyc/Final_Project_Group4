```python
import pandas as pd
```

```python
sales_by_model = "static/data/sales_by_model.csv"
model_sales_df = pd.read_csv(sales_by_model)
```

```python
model_sales_df
```

```python
yearly_ev_sales = "static/data/ev_sales_yearly.csv"
yearly_ev_df = pd.read_csv(yearly_ev_sales)
yearly_ev_df
```

```python
yearly_ev_df2 = yearly_ev_df.rename(columns = {"Unnamed: 0": "Type"})
```

```python
yearly_ev_df_cleaned = yearly_ev_df2.drop(["2000", "2001", "2002", "2003", "2004", "2005",\
                                           "2006", "2007", "2008", "2009", "2010"], axis = 1)
yearly_ev_df_cleaned
```

```python
tesla_sales_df = model_sales_df[model_sales_df.Vehicle.str.contains('tesla',case=False)]
tesla_sales_df
```

```python
tesla_market_share_data = {"Brand": "Tesla",\
                           "2011": tesla_sales_df["2011"].sum()/yearly_ev_df_cleaned["2011"].sum(),\
                          "2012": tesla_sales_df["2012"].sum()/yearly_ev_df_cleaned["2012"].sum(),\
                          "2013": tesla_sales_df["2013"].sum()/yearly_ev_df_cleaned["2013"].sum(),\
                          "2014": tesla_sales_df["2014"].sum()/yearly_ev_df_cleaned["2014"].sum(),\
                          "2015": tesla_sales_df["2015"].sum()/yearly_ev_df_cleaned["2015"].sum(),\
                          "2016": tesla_sales_df["2016"].sum()/yearly_ev_df_cleaned["2016"].sum(),\
                          "2017": tesla_sales_df["2017"].sum()/yearly_ev_df_cleaned["2017"].sum(),\
                          "2018": tesla_sales_df["2018"].sum()/yearly_ev_df_cleaned["2018"].sum(),\
                          "2019": tesla_sales_df["2019"].sum()/yearly_ev_df_cleaned["2019"].sum()}
tesla_market_share_df = pd.DataFrame(tesla_market_share_data, index=[0])
tesla_market_share_df
```
